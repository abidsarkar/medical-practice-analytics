```python
import os
from pathlib import Path
from dotenv import load_dotenv
import psycopg2

# Load the .env file from the root directory
dotenv_path = '../../.env'
load_dotenv(dotenv_path=dotenv_path)

```

```python
db_host = os.getenv("DB_HOST")
db_name = os.getenv("DB_NAME")
db_user = os.getenv("DB_USER")
db_password = os.getenv("DB_PASSWORD")
db_port = os.getenv("DB_PORT")
```

```python
# 5. Connect and execute
try:
    connection = psycopg2.connect(
        host=os.getenv("DB_HOST"),
        database=os.getenv("DB_NAME"),
        user=os.getenv("DB_USER"),
        password=os.getenv("DB_PASSWORD"),
        port=os.getenv("DB_PORT")
    )
    
    cursor = connection.cursor()
    print(f"connection is successful now you can run your query here")

except Exception as error:
    print(f"Error while connecting or executing SQL: {error}")
```

```python
#delete table query
delete_all_table = f"""DROP TABLE IF EXISTS patients,providers,appointments,
encounters,diagnoses,procedures,claims,claim_line_items,payments,
insurance_companies,patient_insurance CASCADE;"""
cursor.execute(delete_all_table)
connection.commit() 
```

```python
if 'cursor' in locals() and cursor:
    cursor.close()
if 'connection' in locals() and connection:
    connection.close()
    print("PostgreSQL connection is closed.")
```
